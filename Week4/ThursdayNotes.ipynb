Using Boolean Masks to find data
=

```python
theta = arange(0.,2*pi, pi/100)
plot(theta, sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
show()
        
```

```python
y = sin(theta[mask])
jj = (y == max(y))
y[jj] = NaN
```

```python
plot(theta, sin(theta))
plot(theta[mask],y, linewidth = 5)
show()
```

Interpolation
=

What if our range is so small that our Boolean mask doesn't find anything? Create a new theta array that doesn't have a value of 0.5. When you try to find it with a mask, all entries in the mask will be false. 

```python
theta = arange(0., 2*pi+0.1, pi/10)
mask = (theta == 0.5)
print mask
```

How can we numerically find a value that is between the existing ones? We'll use interpolation. We have to import the interpolation routines from scripy - the scientific python computing package. 

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta, sin(theta)) #interpolating function created
newSin = sinInterp(newTheta)
print newSin
```

```python
plot(theta, sin(theta), newTheta, newSin, 'r+', markersize = 20, markeredgewidth = 3)
show()
```

other way to interpolate

```python
thetaNew = arange(0,2*pi+0.1,pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
show()
```

```python
plot(thetaNew, sinNew - sin(thetaNew))
ylabel('difference between interp and exact value')
show()
```

```python
sinNewC = interp1d(theta, sin(theta), kind = 'cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew),thetaNew,sinNewC-sin(thetaNew), 'r')
subplot(2,1,2)
plot(thetaNew, sinNewC - sin(thetaNew), 'r')
show()
```
