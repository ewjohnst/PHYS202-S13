Using Boolean Masks to find data
=

```python
theta = arange(0.,2*pi, pi/100)
plot(theta, sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
show()
        
```

```python
y = sin(theta[mask])
jj = (y == max(y))
y[jj] = NaN
```

```python
plot(theta, sin(theta))
plot(theta[mask],y, linewidth = 5)
show()
```

Interpolation
=

What if our range is so small that our Boolean mask doesn't find anything? Create a new theta array that doesn't have a value of 0.5. When you try to find it with a mask, all entries in the mask will be false. 

```python
theta = arange(0., 2*pi+0.1, pi/10)
mask = (theta == 0.5)
print mask
```

How can we numerically find a value that is between the existing ones? We'll use interpolation. We have to import the interpolation routines from scripy - the scientific python computing package. 

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta, sin(theta)) #interpolating function created
newSin = sinInterp(newTheta)
print newSin
```

```python
plot(theta, sin(theta), newTheta, newSin, 'r+', markersize = 20, markeredgewidth = 3)
show()
```

Differentiation
=

```python
x = arange(0.,10.01,0.1)
y = x**5
```

```python
dy = diff(y) #y(n+1)-y(n)
#help(diff)
print len(y), len(dy)
dx = diff(x)
dydx = dy/dx
print len(dydx)
```

```python
#plot(x, dydx)
#explain why that plotting doesn't work, "x and y must have the same first dimension"
```

```python
# specify the size of dy ahead because diff returns an array of n-1 elements
dydx1 = zeros(y.shape,float) #we know it will be this size
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
```

```python
#calculate dydx by center differencing using array slicing
dydx2 = zeros(y.shape,float) # we know it will be this size
dydx2[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2]) #center difference
dydx2[0] = (y[1]-y[0])/(x[1]-x[0]) #forward difference
dydx2[-1] = (y[-1]-y[-2])/(x[-1]-x[-2]) #backward difference
```

```python
dydxExact = 5*x**4
```

```python
plot(x,dydx1,label='forward difference')
plot(x,dydx2,label='center diffrerence')
plot(x,dydxExact,label='exact')
legend(loc='upper left')
show()
#help(plot)
```

```python
percentError1 = 100*abs( dydx1[1:] - dydxExact[1:] )/dydxExact[1:]
percentError2 = 100*abs( dydx2[1:] - dydxExact[1:] )/dydxExact[1:]
#use semilog y - axis to see the variation better
semilogy(x[1:], percentError1, label = 'forward difference')
semilogy(x[1:], percentError2, label = 'center difference')
ylabel("percent error with dydxExact")
legend(loc ="upper right")
show()
```
