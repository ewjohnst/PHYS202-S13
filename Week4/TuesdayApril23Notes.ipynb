## $V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}} - \frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
# create an x-y grid covering the plane -2<x<2 and -2<y<2
dx = 0.06
dy = 0.06
x,y = meshgrid(arange(-2., 2., dx), arange(-2., 2., dy))
#print x.ndim
#print y.shape
```

```python
k = 1; q = 1; d = 1
V = (k*q/sqrt(x**2+(y-d)**2)) - (k*q/sqrt(x**2+(y+d)**2))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.spectral)

cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(x, y, V, zdir='z', offset=-40., cmap=cm.spectral)
cset = ax.contour(x, y, V, zdir='x', offset=-2., cmap=cm.spectral)
cset = ax.contour(x, y, V, zdir='y', offset=-2., cmap=cm.spectral)
```

```python
levels = array((8., 4., 2., 1., 0.5, 0.25, 0., -0.25, -.5, -1., -2., -4., -8.))
contour(x, y, V, levels, cmap=cm.jet)
show()
```
