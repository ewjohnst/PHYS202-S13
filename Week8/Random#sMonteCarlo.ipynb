Cumulative Distribution Funtion (CDF)
-

###Analytic Inverse CDF Transform


Consider $f(x) = \frac{Sin(x)}{2}$ over 0 < x < $\pi$ . it is 0 everywhere else. 
The CDF in this case is just the integral of f(x) below the bound $F(x) = \int^x_{-∞} f(t)dt = cos(0)-\frac{cos(x)}{2}$ for 0 < x < $\pi$ and is 0 everywhere else. 

```python
x = arange(0., pi+0.001, 0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0) - cos(x))/2.
    
plot(x, f(x), label="True Distribution", lw=3, color='blue')
plot(x, F(x), label="Cumulative Function", lw = 4, color='green')

xlim(0, 3.5)
ylim(0, 1.1)

legend(loc=0)
show()
```

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N)

#inverse CDF
Finv = lambda z: arccos(cos(0) - 2*z)
    
#sampled x values
xsamples = Finv(z)

#how do they compare to the original?
hist(xsamples, 50, normed=True, label='Sampled Distribution')
plot(x, f(x), label='True Distribution', lw = 3, color = 'red')

xlim(0, pi+0.1)
ylim(0, 0.7)

legend(loc=0)
show()
```
