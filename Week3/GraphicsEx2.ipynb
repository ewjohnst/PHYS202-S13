```python
import numpy as  np
t = np.arange(0.0, 1.0+0, 0.01)
c = np.cos(2*2*pi*t)
s = np.sin(4*pi*t)
```

```python
fig = plot(t, c, color = "blue", linewidth = 2.5, label = "$cos(4{\pi}t)$")
plot(t, s, color = "red", linewidth = 2.5, label = "$sin(4{\pi}t)$", linestyle = "--")
xlabel("Time(s)", fontsize = 14)
ylabel("Voltage(mV)", fontsize = 14)
# Add title for convenience
title("A Simple Plot", fontsize = 22)
grid(True)
legend(loc = "lower left")
savefig("A Simple Plot.pdf")
```
