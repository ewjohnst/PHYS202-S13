###Line Plots


```python
y = arange(-5, 5.1, 0.1)  #this sets y to an array between -5 and 5 (including both)
E = cosh(y)    #sets the E to the value of the hyperbolic cosine of each value y in our array
p_abs = sinh(y)   #sets the variable called the absolute value of P equal to the hyperbolic sine of each value in our array y
```

```python
figure(figsize=(8,6)) #set figure size in inches

plot(y, E, 'r--', lw=2, label="$E=\cosh(x)$") #plot and label data
plot(y, p_abs, 'b-', lw=2, alpha=.7, label="$|p|=\sinh(x)$")

axis([-2*pi,2*pi,-50,50]) #set axis limits

ax = gca()
ax.spines['left'].set_position('center')
ax.spines['right'].set_position('center')
ax.spines['top'].set_position('center')
ax.spines['bottom'].set_position('center') #draw spines (axes) in the center

xlabel("Independent Variable (units)", position=(.83, 0)) #name and position x-axis label
ylabel("Dependent Variable (units)", position=(0, .8)) #name and position y-axis label

xticks([i for i in arange(-2*pi,2*pi+1,pi/2)],
['$-2\pi$','$-3\pi/2$', '$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$', '$+3\pi/2$', '$+2\pi$']) #format x tick labels
yticks([i*10 for i in range(-5,6)]) #format y tick positions

title("Title of Plot\n", fontsize=20, va="bottom") #set title and font size

grid() #show grid, duh
legend(loc='lower right') #show legend, set location
show() #show all the things!

```

```python
figure=(figsize(8,6))
plot(y,p_abs, color = "red", label = "sinh(x)", lw=2)    #plotting the absolute value of p against y
plot(y, E, color = "blue", label = "cosh(x)", lw=3, linestyle='--')    #plotting E against y
xlabel("y values", fontsize = 14, position=(.84,0))          #making my x-axis label and sizing the font
ylabel("|p|, E", fontsize = 14, position=(0,0.8))            #making my y-axis label and sizing the font
title("|P| and E vs. y")                   #creating the appropriate title and adding the legend in the correct spot
legend( loc = "lower right")
axis([-2*pi,2*pi,-50,50])                  #set axis limits
grid()
ax = gca()                                 #controls axes placement
ax.spines['top'].set_color('none')         #Removing the top, right, and left part of the graph box
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center') #these 2 set axes in center
ax.spines['left'].set_position('center')

ax.yaxis.set_ticks_position('none')        #Removing the tick marks around the bounding box
ax.xaxis.set_ticks_position('none')
xaxis_loc, yaxis_loc = 0, 0

```
