###Line Plots


```python
y = arange(-5, 6, 0.1)  #this sets y to an array between -5 and 5 (including both)
E = cosh(y)    #sets the E to the value of the hyperbolic cosine of each value y in our array
p = sinh(y)   #sets the variable called the absolute value of P equal to the hyperbolic sine of each value in our array y
k = abs(p)           #takes the absolute value of our p function
```

```python
plot(y,k, color = "red", label = "|p|")    #plotting the absolute value of p against y
plot(y, E, color = "blue", label = "E")    #plotting E against y
xlabel("y values", fontsize = 14)          #making my x-axis label and sizing the font
ylabel("|p|, E", fontsize = 14)            #making my y-axis label and sizing the font
title("|P| and E vs. y")                   #creating the appropriate title and adding the legend in the correct spot
legend( loc = "upper left")
ax = gca()                                 #controls axes placement
ax.spines['top'].set_color('none')         #Removing the top, right, and left part of the graph box
ax.spines['right'].set_color('none')

ax.yaxis.set_ticks_position('none')        #Removing the tick marks around the bounding box
ax.xaxis.set_ticks_position('none')
xaxis_loc, yaxis_loc = 0, 0

```
