###Line Plots


```python
y = arange(-5, 5.1, 0.1)  #this sets y to an array between -5 and 5 (including both)
E = cosh(y)    #sets the E to the value of the hyperbolic cosine of each value y in our array
p_abs = sinh(y)   #sets the variable called the absolute value of P equal to the hyperbolic sine of each value in our array y
```

```python
figure=(figsize(8,8))
plot(y,p_abs, color = "red", label = "sinh(x)", lw=2)    #plotting the absolute value of p against y
plot(y, E, color = "blue", label = "cosh(x)", lw=3, linestyle='--')    #plotting E against y
xlabel("x values", fontsize = 14, position=(.84,0))          #making my x-axis label and sizing the font
ylabel("|p|, E", fontsize = 14, position=(0,0.8))           #making my y-axis label and sizing the font
title("|P| and E vs. y")                   #creating the appropriate title and adding the legend in the correct spot
legend( loc = "lower right")
axis([-2*pi,2*pi,-50,50])                  #set axis limits
grid()
ax = gca()                                 #controls axes placement
ax.spines['top'].set_color('none')         #Removing the top, right, and left part of the graph box
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center') #these 2 set axes in center
ax.spines['left'].set_position('center')

ax.yaxis.set_ticks_position('none')        #Removing the tick marks around the bounding box
ax.xaxis.set_ticks_position('none')
xaxis_loc, yaxis_loc = 0, 0


```

####Subplots

```python
#figure(2, figsize=(10.5,5)) #make new figure with index 2

subplot(1,2,1) #In a 1-row, 2-column grid, sets the current subplot to position 1
plot(y, E, 'r-', lw=2)
#xlabel("x values", fontsize = 14)          #making my x-axis label and sizing the font
#ylabel("cosh(x)", fontsize = 14) #any formatting done here will only apply to this subplot:
title("Subplot 1")

subplot(1,2,2) #In a 1-row, 2-column grid, sets the current subplot to position 2 (shorthand form)
plot(y, p_abs, 'r-', lw=2)
#axis([-5,5,-10,10])           #xlabel('x values', fontsize=14)  #insert nice formatting here, as above
#ylabel('sinh(x)', fontsize=14) #any formatting done here will only apply to this subplot:
title("Subplot 2")

subplot(1,2,2) #I can switch the active subplot at any time if I need to
axis([-5,5,-40,40])

show()

```

```python
#figure(2, figsize=(10.5,5)) #make new figure with index 2

subplot(2,1,1) #In a 1-row, 2-column grid, sets the current subplot to position 1
plot(y, E, 'r-', lw=2)
#any formatting done here will only apply to this subplot:
title("Subplot 1")

subplot(2,1,2) #In a 1-row, 2-column grid, sets the current subplot to position 2 (shorthand form)
plot(y, p_abs, 'r-', lw=2)
#any formatting done here will only apply to this subplot:
title("Subplot 2")

subplot(2,1,1) #I can switch the active subplot at any time if I need to
axis([-5,5,-1,20])

show()
```

```python
#Basic functions
x = arange(-5,5,0.1)
y = cosh(x)
z = sinh(x)
```

```python
#subplots 
subplot(1,2,1)
pylab.scatter(x, y, color='red', s=.2) # or just plot(<stuff goes here>)

subplot(1,2,2)
pylab.scatter(x,z, color='blue', s=.2)
```

```python
#Formatting Stuff

#plotting, if you put a second plot below first they will be on same graph
plot(x, y, lw = 1., label = "<what's being plotted>", linestyle = '--', color = 'blue')

#title
ax.set_title('<title name>')

#setting axes (called spines)
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')

#tick mark positioning
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#axes names
xlabel("<xlabel name>", fontsize=14)
ylabel("<ylabel name>", fontsize=14)

#legend location
plt.legend(loc='upper left', fontsize = 8)

#gets rid of stuff right before graph that pops up
show()
```
