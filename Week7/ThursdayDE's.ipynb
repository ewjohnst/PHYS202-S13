Use Euler's method to solve the following first-order DE with the given condition y(0) = 0 and
$\frac{dy}{dx}=x+2y$. 
Find the value of y when x = 1, that is y(1) using a step size of 0.25 and then 0.02.

```python
f = lambda x, y: x + 2*y

def euler(dydx, y0, stepsize):
    x = np.arange(0, 1+stepsize, stepsize)
    y = np.zeros(x.size)
    
    y[0] = y0
    for i in range(y.size -1):
        y[i+1] = y[i] + dydx(x[i], y[i])*stepsize          
    return x,y

ans = euler(f, 0, .25)[1][-1]
print ans
ans = euler(f, 0, .02)[1][-1]
print ans

```

Plot the results from part A for the two different step sizes as well as the exact solution $y=0.25e^{2x}-0.5x-0.25$. Include a legend. Which step size gives the move accurate result?

```python
plot(euler(f, 0, .25)[0] ,euler(f, 0, .25)[1], 'red', label='Step size of .25')
plot(euler(f, 0, 0.02)[0], euler(f, 0, 0.02)[1], 'blue', label='Step size of .02')
x = np.arange(0, 1.01, 0.01)
plot(x, 0.25*e**(2*x)-0.5*x-0.25, 'green', label='Exact solution')
legend(loc='upper left')
show()
```

The smaller step size gives a much more accurate result, which I expected.
