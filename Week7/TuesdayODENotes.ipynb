Approximate Solutions

```python
g = 9.8 #m/s^2
l = 0.5 #m
th0 = 20.*pi/180 # 20 degrees, but need be in radians.

t = arange(0.,10.01,0.01)
theta_app = lambda t: th0*cos(sqrt(g/l)*t)

plot(t,theta_app(t))
show()
```

```python
def deriv(theta_current,t): # Return derivatives of the array theta
    g = 9.8
    l = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta,domega])
```

```python
from scipy.integrate import odeint
#odeint?
```

```python
t = arange(0.,10.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = array([th0,omega0]) # initial values, th0 from before...

theta = odeint(deriv,theta_init,t) # pass the function deriv to the solver...
```

```python
theta.shape
```

```python
plot(t,theta_app(t),"b-", t,theta[:,0],"r--")
show()
```

```python
th0 = 50.*pi/180 # 50 degrees

theta_init = array([th0,omega0]) # initial values
theta = odeint(deriv,theta_init,t)

plot(t,theta_app(t),"b-",t,theta[:,0],"r--")
show()
```

1: see what happens to the period of the pendulum as you change the initial angle of release. Plot θ  vs t  for 5 different initial angles, θ 0   , spanning 0 to π  . (i.e. 0, π /4 , π /2 , 3π /4 , and π  ). Put them alll on the same plot. What happens to the period of the pendulum as the initial angle increases from π /4 to 3π /4 ? What happens to the trajectory when the initial angle is 0 and π  ?

```python
figure(figsize=(10,8))

th0 = 0 
theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)
plot(t,theta[:,0],label=r"$\theta$ = 0")

th0 = pi/4 
theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)
plot(t,theta[:,0],label=r"$\theta$ = $\frac{\pi}{4}$")

th0 = pi/2 
theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)
plot(t,theta[:,0],label=r"$\theta$ = $\frac{\pi}{2}$")

th0 = 3.*pi/4 
theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)
plot(t,theta[:,0],label=r"$\theta$ = $3\frac{\pi}{4}$")

th0 = pi 
theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)
plot(t,theta[:,0],label=r"$\theta$ = $\pi$")

legend(loc="lower right")
show()
```

As theta increases from π /4 to 3π /4, the amplitutde and period increase.
