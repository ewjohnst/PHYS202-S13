Approximate Solution
-

```python
g = 9.8 #m/s^2
l = 0.5 #m
theta0 = 20.*pi/180 # 20 degrees but must use radians
t = arange(0., 10.+0.01, 0.01)
theta_app = lambda t: theta0*cos(sqrt(g/l)*t)
    
plot (t, theta_app(t))
show()
```

Numerical Solution
-

```python
def deriv(theta_current, t): #returns the derivative of
    g = 9.8 #m/s^2           #array theta
    l = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta, domega])
```

```python
from scipy.integrate import odeint
#odeint??
```

```python
t = arange(0., 10.+0.01, 0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = array([theta0, omega0]) #intiial values, theta0 from before

theta = odeint(deriv, theta_init, t) #pass the funtion deriv to the solver
```

```python
theta.shape
```

```python
plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
show()
```

```python
theta0 = 50.*pi/180 #50 degrees

theta_init = array([theta0, omega0]) #initial values
theta = odeint(deriv,theta_init, t)

plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
```

Put them all on the same plot. What happens to the period of the pendulum as the initial angles increases from $\pi$ /4 to 3$\pi$ /4? What happens to the trajetory when the initial angle is 0 and $\pi$?


```python
figure(figsize=(10,8))
theta0 = 0.
theta_init = array([theta0, omega0]) #initial values
theta = odeint(deriv,theta_init, t)
plot(t, theta[:,0], 'red', label=r'$\theta$ = 0')
legend('lower right')

theta0 = pi/4
theta_init = array([theta0, omega0]) #initial values
theta = odeint(deriv,theta_init, t)
plot(t, theta[:,0], 'blue',label=r'$\theta$ = $\pi$/4')
legend('lower right')

theta0 = pi/2
theta_init = array([theta0, omega0]) #initial values
theta = odeint(deriv,theta_init, t)
plot(t, theta[:,0], 'green',label=r'$\theta$ = $\pi$/2')
legend('lower right')

theta0 = 3*pi/4
theta_init = array([theta0, omega0]) #initial values
theta = odeint(deriv,theta_init, t)
plot(t, theta[:,0], 'purple',label=r'$\theta$ = 3$\pi$/4')
legend('lower right')

theta0 = pi
theta_init = array([theta0, omega0]) #initial values
theta = odeint(deriv,theta_init, t)
plot(t, theta[:,0], 'black', label=r'$\pi$')
legend(loc='lower right')
```
